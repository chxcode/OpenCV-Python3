openCv æä¾›äº†ä¸¤ç§å˜æ¢å‡½æ•°ï¼Œcv2.warpAffine å’Œ cv2.warpPerspectiveã€‚é€šè¿‡è¿™ä¸¤ä¸ªå‡½æ•°ï¼Œä½ å¯ä»¥å®ç°æ‰€æœ‰çš„å›¾åƒå˜æ¢ã€‚cv2.warpAffineæ¥å—çš„å‚æ•°æ˜¯2x3çš„å˜æ¢çŸ©é˜µã€‚cv2.warpPerspective æ¥å—çš„å‚æ•°æ˜¯3x3çš„å˜æ¢çŸ©é˜µã€‚

###  æ‰©å±•ç¼©æ”¾

æ‰©å±•ç¼©æ”¾åªæ˜¯æ”¹å˜å›¾åƒçš„å°ºå¯¸å¤§å°ã€‚OpenCV æä¾›çš„å‡½æ•° cv2.resize() å¯ä»¥å®ç°è¿™ä¸ªåŠŸèƒ½ã€‚å›¾åƒçš„å°ºå¯¸å¯ä»¥è‡ªå·±æ‰‹åŠ¨è®¾ç½®ï¼Œä½ ä¹Ÿå¯ä»¥æŒ‡å®šç¼©æ”¾å› å­ã€‚æˆ‘ä»¬å¯ä»¥é€‰æ‹©ä½¿ç”¨ä¸åŒçš„æ’å€¼æ–¹æ³•ã€‚åœ¨ç¼©æ”¾æ—¶æˆ‘ä»¬æ¨èä½¿ç”¨ cv2.INTER_AREA ã€‚åœ¨æ‰©å±•æ—¶æˆ‘ä»¬æ¨èä½¿ç”¨ v2.INTER_CUBICï¼ˆæ…¢) å’Œ v2.INTER_LINEARã€‚é»˜è®¤æƒ…å†µä¸‹æ‰€æœ‰æ”¹å˜å›¾åƒå°ºå¯¸å¤§å°çš„æ“ä½œä½¿ç”¨çš„æ’å€¼æ–¹æ³•éƒ½æ˜¯ cv2.INTER_LINEARã€‚ ä½ å¯ä»¥ä½¿ç”¨ä¸‹é¢ä»»æ„ä¸€ç§æ–¹æ³•æ”¹å˜å›¾åƒçš„å°ºå¯¸ï¼š

```python
import cv2
import numpy as np

img = cv2.imread('./image/messi5.jpg')

res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)

#æˆ–è€…

height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

while(1):
    cv2.imshow('res',res)
    cv2.imshow('img',img)
    if cv2.waitKey(1) & 0xFF == 27: 
        break
cv2.destroyAllWindows()
```

### å¹³ç§»

å¹³ç§»æ˜¯è½¬æ¢ä¸€ä¸ªå¯¹è±¡çš„ä½ç½®ã€‚å¦‚æœä½ æ²¿(x ,y) æ–¹å‘ç§»åŠ¨,ç§»åŠ¨çš„è·ç¦»æ˜¯$$({t_x},{t_y})$$ä½ å¯ä»¥ç”¨ä¸‹é¢çš„æ–¹å¼æ„å»ºç§»åŠ¨çŸ©é˜µï¼šM = $ \begin{bmatrix} 1 & 0 & t_x \\ 0 & 1 & t_y \\ \end{bmatrix} $ä½ å¯ä»¥ä½¿ç”¨ Numpy æ•°ç»„æ„å»ºä¸€ä¸ªçŸ©é˜µæ•°æ®ç±»å‹æ˜¯ np.float32 ï¼Œç„¶åæŠŠå®ƒä¼ ç»™å‡½æ•° cv2.warpAffine()ã€‚çœ‹ä¸‹é¢çš„è¿™ä¸ªä¾‹å­ï¼Œå®ƒç§»åŠ¨äº†(100,50)ä¸ªåƒç´ ã€‚

```python
import cv2
import numpy as np

img = cv2.imread('image/messi5.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### æ—‹è½¬


å¯¹ä¸€ä¸ªå›¾åƒæ—‹è§’åº¦ Î¸,éœ€è¦ä½¿ç”¨åˆ°ä¸‹é¢å½¢å¼çš„æ—‹è½¬çŸ©é˜µï¼šM = $ \begin{bmatrix} cosğ›³ & -sinğ›³ \\ sinğ›³ & cosğ›³ \\ \end{bmatrix} $ ä½†æ˜¯OpenCVæä¾›äº†å¯è°ƒæ—‹è½¬ä¸­å¿ƒçš„ç¼©æ”¾æ—‹è½¬ï¼Œè¿™æ ·ä½ å¯ä»¥åœ¨ä»»ä½•ä½ å–œæ¬¢çš„ä½ç½®æ—‹è½¬ã€‚ä¿®æ­£åçš„å˜æ¢çŸ©é˜µä¸º:$ \begin{bmatrix} Ã¥ & ÃŸ & (1-Ã¥)Â·center.x - ÃŸÂ·center.y \\ -ÃŸ & Ã¥ & ÃŸÂ·center.x + (1-Ã¥)center.y \\ \end{bmatrix} $å…¶ä¸­ï¼ŒÃ¥ = scaleÂ·cosğ›³ï¼ŒÃŸ = scaleÂ·sinğ›³

ä¸ºäº†æ„å»ºè¿™ä¸ªæ—‹è½¬çŸ©é˜µ  OpenCV æä¾›äº†ä¸€ä¸ªå‡½æ•° cv2.getRotationMatrix2Dã€‚ ä¸‹é¢çš„ä¾‹å­æ˜¯åœ¨ä¸ç¼©æ”¾çš„æƒ…å†µä¸‹å°†å›¾åƒæ—‹  90 åº¦ã€‚

```python
img = cv2.imread('messi5.jpg',0)
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

### é€è§†å˜æ¢

å¯¹äºé€è§†å˜æ¢ï¼Œæˆ‘ä»¬éœ€è¦ä¸€ä¸ª 3X3 å˜æ¢çŸ©é˜µã€‚åœ¨å˜æ¢å‰åç›´çº¿è¿˜æ˜¯ç›´çº¿ã€‚è¦æ„å»ºè¿™ä¸ªå˜æ¢çŸ©ï¼Œä½ éœ€è¦åœ¨åœ¨è¾“å…¥å›¾åƒä¸Šæ‰¾ 4 ä¸ªç‚¹ï¼Œä»¥åŠä»–ä»¬åœ¨è¾“å‡ºå›¾åƒä¸Šå¯¹åº”çš„ä½ç½®ã€‚ å››ä¸ªç‚¹ä¸­çš„ä»»æ„ä¸‰ä¸ªä¸èƒ½å…±çº¿ã€‚è¿™ä¸ªå˜æ¢çŸ©é˜µå¯ä»¥ç”±å‡½æ•° cv2.getPerspectiveTransform() æ„å»ºã€‚ç„¶åæŠŠè¿™ä¸ªçŸ©é˜µä¼ ç»™å‡½æ•° cv2.warpPerspectiveã€‚

ä»£ç å¦‚ä¸‹ï¼š

```python


img = cv2.imread('sudokusmall.png')
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
